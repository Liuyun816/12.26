```python
from tkinter import *
from tkinter import ttk
from enum import Enum

class State(Enum):
    calculated = 0
    in_number = 1
    post_operator = 2
    
class Row(Frame):
    def __init__(self,parent,buttons=None, *args, **kwargs):
        super().__init__( parent,*args, **kwargs)
        self.pack(side=TOP, expand=True, fill=BOTH)
        for button in buttons:
            ttk.Button(self,text = button[0],command=button[1])\
                .pack(side=LEFT, expand=True, fill=BOTH)
    
    def add_button(self,  text=None, command=None):
        button = ttk.Button(self, text=text, command=command)
        button.pack(side=LEFT, expand=True, fill=BOTH)
        return self
        
class Calculator(Tk):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.setup_ui()
        self.reset()
      
    def setup_ui(self):
        self.title('Simple calculator')
        self.geometry('400x600')
        
        self.result_string = StringVar()
        self.result_string.set('0')
        
        # Output
        output = ttk.Label(self, textvariable=self.result_string,
                           font=('TKDefaulFont', 48), anchor=E)
        output.pack(side=TOP, expand=True, fill=BOTH)
        
        # row 1 
        Row(self,[('CE',self.on_ce_click),
                 ('C', self.on_c_click),
                 ('Back', self.on_back_click),
                 ('/',self.gen_on_operator_click('/'))])    
        #self.add_button(row, 'CE', self.on_ce_click)
      
        # row 2 
        Row(self,[( '7', self.gen_on_number_click('7')),
                 ('8', self.gen_on_number_click('8')),
                 ('9', self.gen_on_number_click('9')),
                 ('x', self.gen_on_operator_click('x'))])

        # row 3 
        Row(self,[('4', self.gen_on_number_click('4')),
                 ('5', self.gen_on_number_click('5')),
                 ('6', self.gen_on_number_click('6')),
                 ('-', self.gen_on_operator_click('-'))])

        # row 4 
        Row(self,[( '1', self.gen_on_number_click('1'),)
                  ('2', self.gen_on_number_click('2')),
                  ('3', self.gen_on_number_click('3')),
                  ('+', self.gen_on_operator_click('+'))])
#        self.add_button(row, '1', self.on_1_click)

        # row 5 
        Row(self,[( '+/-', self.on_negative_click),
                 '0', self.gen_on_number_click('0'),
                 '.', self.gen_on_number_click('.'),
                 ('=', self.gen_on_operator_click('='))])
                
        
    def on_ce_click(self):
        print('ce clicked.')
        if self.state is State.in_number and len(self.number2) > 1:
            self.number2 = ''
        self.result_string.set('0')
        
    def on_c_click(self):
        print('c clicked.')
        self.reset()
        
    def reset(self):
        self.number1 = 0
        self.number2 = None
        self.operator = '+' #None
        self.state = State.calculated
        self.result_string.set('0')

    def on_back_click(self):
        print('back clicked.')
        if self.state is State.in_number :
            if len(self.number2) >= 1:
                self.number2 = self.number2[:-1]
            self.result_string.set(self.number2 or '0')
                

    def on_division_click(self):
        self.on_operator('/')
        print('division clicked.')

    def gen_on_number_click(self, ch):
        return lambda: self.on_number(ch)
    
    def gen_on_operatoe_click(self,ch):
        return lambda:self.on_operator(ch)
#    def on_1_click(self):
#        self.on_number('1')
#        print('1 clicked.')

    def is_valid(self,number_string):
        if number_string.startswith('00'):
            return False
        try:
            float(number_string)
            return True
        except:
            return False
    def on_number(self, ch):
        print(self.state)
        if self.state == State.in_number:
            if self.is_valid(self.number2 + ch):
                self.number2 = self.number2 + ch
        else: 
            self.number2 = ch
            self.state = State.in_number
            
        self.result_string.set(self.number2)    
   
    def on_multiply_click(self):
        print('x clicked.')
        self.on_operator('*')
    
    def on_minus_click(self):
        print('- clicked.')
        self.on_operator('-')

    def on_add_click(self):
        print('+ clicked.')
        self.on_operator('+')
    
    def on_operator(self, operator):
        print('{}',state)
        if self.state == State.calculated:
            self.state = State.post_operator
        elif self.state == State.in_number1:
            self.state = State.post_operator
            self.on_equal_click()
        else: assert(0)        

        self.operator = operator
        self.state = State.post_operator
   
    def on_equal_click(self):        
        print('= clicked.')
        if self.operator == '+':
            self.number1 = self.number1 + float(self.number2)
        elif self.operator == '-':
            self.number1 = self.number1 - float(self.number2)
        elif self.operator == '*':
            self.number1 = self.number1 * float(self.number2)
        elif self.operator == '/':
            if float(self.number2)==0:
                self.reset()
                self.result_string.set('Divided by 0.')
            else:
                self.number1 = self.number1 / float(self.number2)
#       elif self.operator == None:
#            self.number1 = float(self.number2)

        output = str(self.number1)
        if int(self.number1)== self.number1:
            output = str(int(self.number1))
            
        self.result_string.set(self.number1)
        self.state = State.calculated

    def on_negative_click(self):
        print('+/- clicked.')
        if self.number2.startswith('-'):
            self.number2 = self.number[1:]
        else:
            self.number2 = '-' + self.number2
            
        self.state = State.in_number
        self.result_string
        
if __name__ == '__main__':
    app = Calculator()
    app.mainloop()
    #Calculator.mianloop()
```
