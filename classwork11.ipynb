```python
from tkinter import *
from tkinter import ttk
from enum import Enum

class State(Enum):
    calculated = 0
    in_number1 = 1
    post_operator = 2
    in_number2 = 3
    
class Calculator(Tk):
    def __init__(self,*args,**kwargs):  ## *在函数调用时，调用所有用户没有指定keywords的参数
        super().__init__(*args,**kwargs)  ## **调用指定了keywords的参数
        self.title('Simple calculator')
        self.geometry('400x600')
        
        self.result_string=StringVar()
        self.result_string.set('0')
        self.number1 = '0'
        self.number2 = None
        self.operator = None
        self.state = State.calculated
        
        # Output
        output = ttk.Label(self,textvariable =self.result_string,
                           font = ('TKDefaulFont',48),anchor = E)   #font表示字体，E表示东边（右边）对齐
        output.pack(side = TOP,expand = True,fill = BOTH)
        
        
        # row 1
        row = Frame(self)    #self表示Tk主窗口里
        row.pack(side = TOP,expand = True,fill = BOTH)
        self.add_button(row,'CE',self.on_ce_click)
        self.add_button(row,'C',self.on_c_click)
        self.add_button(row,'Back',self.on_back_click)
        self.add_button(row,'/',self.on_division_click)
        
        # row 2
        row = Frame(self)    
        row.pack(side = TOP,expand = True,fill = BOTH)
        self.add_button(row,'7',self.on_7_click)
        self.add_button(row,'8',self.on_8_click)
        self.add_button(row,'9',self.on_9_click)
        self.add_button(row,'x',self.on_x_click)
        
        # row 3
        row = Frame(self)    
        row.pack(side = TOP,expand = True,fill = BOTH)
        self.add_button(row,'4',self.on_4_click)
        self.add_button(row,'5',self.on_5_click)
        self.add_button(row,'6',self.on_6_click)
        self.add_button(row,'-',self.on_minus_click)
        
        # row 4
        row = Frame(self)    
        row.pack(side = TOP,expand = True,fill = BOTH)
#       self.add_button(row,'1',self.on_1_click)
        self.add_button(row,'1',self.gen_on_number_click('1'))
        self.add_button(row,'2',self.on_2_click)
        self.add_button(row,'3',self.on_3_click)
        self.add_button(row,'+',self.on_add_click)
        
        # row 5
        row = Frame(self)    
        row.pack(side = TOP,expand = True,fill = BOTH)
        self.add_button(row,'+/-',self.on_negative_click)
        self.add_button(row,'0',self.on_0_click)
        self.add_button(row,'.',self.on_dot_click)
        self.add_button(row,'=',self.on_equal_click)
        
    def add_button(self,parent,text = None,command = None):
        button = ttk.Button(parent,text = text,command = command)
        button.pack(side = LEFT,expand = True, fill = BOTH)
        
    def on_ce_click(self):
        print('ce clicked.')
        self.on_operator('ce')
        
    def on_c_click(self):
        print('c clicked.')
        self.on_operator('c')
        
    def on_back_click(self):
        print('back clicked.')
        self.on_operator('back')
        
    def on_division_click(self):
        print('/ clicked.')
        self.on_operator('/')
        
    def on_7_click(self):
        print('7 clicked.')
        self.on_number('7')
        
    def on_8_click(self):
        print('8 clicked.')
        self.on_number('8')
        
    def on_9_click(self):
        print('9 clicked.')
        self.on_number('9')
        
    def on_x_click(self):
        print('x clicked.')
        self.on_operator('x')
        
    def on_4_click(self):
        print('4 clicked.')
        self.on_number('4')
        
    def on_5_click(self):
        print('5 clicked.')        
        self.on_number('5')
        
    def on_6_click(self):
        print('6 clicked.')
        self.on_number('6')
        
    def on_minus_click(self):
        print('- clicked.')
        self.on_operator('-')
        
#    def on_1_click(self):
#        print('1 clicked.')
#        self.on_number('1')
    
    def on_2_click(self):
        print('2 clicked.')
        self.on_number('2')
        
    def on_3_click(self):
        print('3 clicked.')
        self.on_number('3')
        
    def on_add_click(self):
        print('+ clicked.')
        self.on_operator('+')
        
    def on_negative_click(self):
        print('+/- clicked.')
        self.on_operator('+/-')
        
    def on_0_click(self):
        print('0 clicked.')
        
    def on_dot_click(self):
        print('. clicked.')
        self.on_number('.')
        
    def on_equal_click(self):
        print('= clicked.')
        self.on_operator('=')
        if self.operator == '+':
            self.number1 = str(float(self.number1) + float(self.number2))
        elif self.operator == '-':
            self.number1 = str(float(self.number1) - float(self.number2))
        elif self.operator == '*':
            self.number1 = str(float(self.number1) * float(self.number2))     
        elif self.operator == '/':
            self.number1 = str(float(self.number1) / float(self.number2))
            
        self.result_string.set(self.number1)
        self.state = State.calculated
        
    def on_number(self,ch):
        if self.state ==State.calculated:
            self.number1 = ch
            self.state = State.in_number1
            self.result_string.set(self.number1)
        elif self.state == State.in_number1:    
            self.number1 = self.number1 + ch
        elif self.state == State.post_operator:
            self.number2 = ch
            self.state = State.in_number2
        elif self.state ==State.in_number2:
            self.number2 = self.number2 + ch
            self.result_string.set(self.number2)
        else:assert(0)
            
            
    def on_operator(self,operator):
        if self.state == State.calculated:
            self.operator = operator
            self.state = State.post_operator:
        elif self.state ==State.in_number1:
            self.state = State.post_operator
       # elif self.state == State.post_operator:
        elif self.state == State.in_number2:
            self.on_equal_click()
            self.state = State.post_operator
        else: assert(0)    
        
        self.operator = operator
        
if __name__ == '__main__':    ##如果被直接执行而非调用进来
    app = Calculator()
    app.mainloop()
```

```python
def gen_on_number_click(ch):
    return lambda:on_number(ch)       
```
